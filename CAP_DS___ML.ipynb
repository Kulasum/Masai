```python
import pandas as pd

# Assuming df_cleaned is already defined
# Define the function to remove outliers
def remove_outlier(df):
    numerical_cols = df.select_dtypes(include='number')
    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.3 * IQR
        upper_bound = Q3 + 1.3 * IQR  # Ensure correct upper bound
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Remove outliers from df_cleaned
df_cleaned = remove_outlier(df_cleaned)

# Print the cleaned DataFrame
print(df_cleaned)


```

This metric gives the average squared difference between the actual values and the predicted values. A lower MSE indicates better model performance. However, since MSE is based on squared units, it is less interpretable in terms of the original data. In this case, the MSE is quite high, indicating that there might be significant errors in predictions. Root Mean Squared Error (RMSE): 3,286.20

RMSE is the square root of MSE and gives an error measure in the same units as the target variable. The RMSE of 3,286.20 means that, on average, the predictions differ from the actual values by about 3,286 units. This is a direct indication of how far off the predictions are from the actual values. R-squared Score: 0.8287

The R-squared value indicates the proportion of variance in the dependent variable that is predictable from the independent variables. An R-squared value of 0.8287 means that 82.87% of the variance in the target variable is explained by the model. This suggests that the model fits the data well, but it's not perfect. Adjusted R-squared Score: 0.8213

The adjusted R-squared takes into account the number of predictors in the model and adjusts for the possibility of overfitting. Since it is very close to the R-squared score (0.8213 vs. 0.8287), it suggests that the model is not overfitting, and the predictors chosen are relevant and useful. Performance Interpretation: The model has a reasonably good fit, as indicated by the high R-squared and adjusted R-squared values. Over 82% of the variance in the data is explained by the model, which suggests a strong relationship between the predictors and the target. However, the relatively high RMSE (3,286) shows that the predictions might have a significant amount of error in absolute terms. This could indicate that while the model captures the overall trend well, there are outliers or variability in the data that the model doesn't account for well.
